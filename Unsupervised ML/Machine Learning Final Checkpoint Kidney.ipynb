```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X = data.loc[:, data.columns != 'class']
y = data['class']
```

```python
min_max = MinMaxScaler()
min_max.fit(X)
X = min_max.transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics, preprocessing
```

## Supervised Learning

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state=123)
LL = LogisticRegression(max_iter=3000)
LL.fit(X_train,Y_train)
LogR_y_pred_test = LL.predict(X_test)
```

```python
pickle.dump(LL, open('kidney_LL.pkl', 'wb'))
```

```python
print("MSE : ", metrics.mean_squared_error(Y_test,LogR_y_pred_test))
print('R Square Error : ',metrics.r2_score(Y_test,LogR_y_pred_test))
print('Logestic Regression accuracy : ',metrics.accuracy_score(Y_test,LogR_y_pred_test))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

residuals = Y_test - LogR_y_pred_test
sns.displot(residuals, kind='kde')
```

```python
cm = metrics.confusion_matrix(Y_test, LogR_y_pred_test)
sns.heatmap(cm, annot=True, fmt='', cmap='Blues')
```

### KNN

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
accs = []
k_grid = range(1,81,1)
for k in k_grid:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, Y_train)
    Y_pred = knn.predict(X_test)
    accs.append(metrics.accuracy_score(Y_test,Y_pred))
```

```python
# Visualize.
plt.scatter(k_grid,accs,c='red',marker='o',s=10,alpha=0.6)
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('Accuracy vs k')
plt.show()
```

```python
k_grid = np.arange(1,40,1)
parameters = {'n_neighbors':k_grid}
# Optimize the k.
gridCV = GridSearchCV(KNeighborsClassifier(), parameters, cv=2, n_jobs = -1)       # "n_jobs = -1" means "use all the CPU cores".
gridCV.fit(X_train, Y_train)
best_k = gridCV.best_params_['n_neighbors']
print("Best k : " + str(best_k))
```

```python
# Test with the best k.
KNN_best = KNeighborsClassifier(n_neighbors=best_k)
KNN_best.fit(X_train, Y_train)
knn_y_pred = KNN_best.predict(X_test)
print( "Best Accuracy : " + str(np.round(metrics.accuracy_score(Y_test,knn_y_pred),3)))
```

```python
pickle.dump(KNN_best, open('kidney_KNN.pkl', 'wb'))
```

### K-Means

```python
from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
kmeans=KMeans(n_init=10,n_clusters=5, random_state=0)  #Cluster our data by choosing 5 as number of clusters
kmeans.fit(data)
```

```python
#This instruction enables us to print the label of each point in our data after the clustering is done.
labels=pd.DataFrame(kmeans.labels_)
labels
```

```python
kmeans.predict(data)
print(kmeans.cluster_centers_)   #Printing the coordinates of cluster centers.
```

```python
data['label']=labels
```

```python
#K-means plot Each colored group represents a cluster.


plt.scatter(data["serum creatinine"][data.label == 0],          
            data["blood urea"][data.label == 0],s=80,c='magenta',label='Careful')
plt.scatter(data["blood urea"][data.label == 1],
           data["serum creatinine"][data.label == 1],s=80,c='yellow',label='Standard')
plt.scatter(data["blood urea"][data.label == 2],
           data["serum creatinine"][data.label == 2],s=80,c='green',label='Target')
plt.scatter(data["blood urea"][data.label == 3], 
           data["serum creatinine"][data.label == 3],s=80,c='cyan',label='Careless')
plt.scatter(data["blood urea"][data.label == 4], 
           data["serum creatinine"][data.label == 4],s=80,c='burlywood',label='Sensible')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('Blood urea')
plt.ylabel('serum creatinine')
plt.legend()
plt.show()
```
