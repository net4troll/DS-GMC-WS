```python
for i in range(data.shape[0]):
    if data.iloc[i,25]=='ckd\t':
        data.iloc[i,25]='ckd'
    if data.iloc[i,20] in [' yes','\tyes']:
        data.iloc[i,20]='yes'
    if data.iloc[i,20]=='\tno':
        data.iloc[i,20]='no'
    if data.iloc[i,21]=='\tno':
        data.iloc[i,21]='no'
    if data.iloc[i,16]=='\t?':
        data.iloc[i,16]=np.nan
    if data.iloc[i,16]=='\t43':
        data.iloc[i,16]='43'
    if data.iloc[i,17]=='\t?':
        data.iloc[i,17]=np.nan
    if data.iloc[i,17]=='\t6200':
        data.iloc[i,17]= '6200'
    if data.iloc[i,17]=='\t8400':
        data.iloc[i,17]= '6200'
    if data.iloc[i,18]=='\t?':
        data.iloc[i,18]=np.nan
    if data.iloc[i,25]=='ckd':
        data.iloc[i,25]='yes'
    if data.iloc[i,25]=='notckd':
        data.iloc[i,25]='no'
```

```python
def fillna_imputer(l):
    from sklearn.impute import KNNImputer
    imputer = KNNImputer(n_neighbors=5)
    for i in l :
        data[i] = imputer.fit_transform(data[[i]])
```

```python
data = data.drop('id',axis=1)
```

```python
data['class'].value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X = data.loc[:, data.columns != 'class']
y = data['class']
```

```python
min_max = MinMaxScaler()
min_max.fit(X)
X = min_max.transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics, preprocessing
```

## Supervised Learning

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2, random_state=123)
LL = LogisticRegression(max_iter=3000)
LL.fit(X_train,Y_train)
LogR_y_pred_test = LL.predict(X_test)
```

```python
print("MSE : ", metrics.mean_squared_error(Y_test,LogR_y_pred_test))
print('R Square Error : ',metrics.r2_score(Y_test,LogR_y_pred_test))
print('Logestic Regression accuracy : ',metrics.accuracy_score(Y_test,LogR_y_pred_test))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

residuals = Y_test - LogR_y_pred_test
sns.displot(residuals, kind='kde')
```

```python
cm = metrics.confusion_matrix(Y_test, LogR_y_pred_test)
sns.heatmap(cm, annot=True, fmt='', cmap='Blues')
```

### KNN

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
```

```python
accs = []
k_grid = range(1,81,1)
for k in k_grid:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, Y_train)
    Y_pred = knn.predict(X_test)
    accs.append(metrics.accuracy_score(Y_test,Y_pred))
```

```python
# Visualize.
plt.scatter(k_grid,accs,c='red',marker='o',s=10,alpha=0.6)
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.title('Accuracy vs k')
plt.show()
```

```python
k_grid = np.arange(1,40,1)
parameters = {'n_neighbors':k_grid}
# Optimize the k.
gridCV = GridSearchCV(KNeighborsClassifier(), parameters, cv=2, n_jobs = -1)       # "n_jobs = -1" means "use all the CPU cores".
gridCV.fit(X_train, Y_train)
best_k = gridCV.best_params_['n_neighbors']
print("Best k : " + str(best_k))
```

```python
# Test with the best k.
KNN_best = KNeighborsClassifier(n_neighbors=best_k)
KNN_best.fit(X_train, Y_train)
knn_y_pred = KNN_best.predict(X_test)
print( "Best Accuracy : " + str(np.round(metrics.accuracy_score(Y_test,knn_y_pred),3)))
```

### K-Means

```python
from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
kmeans=KMeans(n_init=10,n_clusters=5, random_state=0)  #Cluster our data by choosing 5 as number of clusters
kmeans.fit(data)
```

```python
#This instruction enables us to print the label of each point in our data after the clustering is done.
labels=pd.DataFrame(kmeans.labels_)
labels
```

```python
kmeans.predict(data)
print(kmeans.cluster_centers_)   #Printing the coordinates of cluster centers.
```

```python
data['label']=labels
```
