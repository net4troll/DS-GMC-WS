```python
import requests
import pandas as pd
import json
u = 'https://storage.googleapis.com/public-sentiments-v9/fxcs-sentiments.json'
res = requests.get(u)
data = json.loads(res.text)
df = pd.DataFrame(data['sentiments'])
sentiment=df[['instrument', 'contrarian_indicator_signal',
       'short_percentage', 'long_percentage', 'contrarian_indicator_type',
       'contrarian_indicator_strength']]
```

```python
from sqlalchemy import create_engine
sqlEngine       = create_engine('mysql+pymysql://root:ouvrestoi@127.0.0.1/dashboard', pool_recycle=3600)
dbConnection    = sqlEngine.connect()
tableName   = "sentiment"
```

```python
try:
    frame = sentiment.to_sql(tableName, dbConnection, if_exists='append',method='multi');
except ValueError as vx:
    print(vx)
except Exception as ex:
    print(ex)
else:
    print(f"Table {tableName} created successfully.");
finally:
    dbConnection.close()
```
